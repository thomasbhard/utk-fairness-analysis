```python
import pandas as pd
import matplotlib.pyplot as plt
from utils import group_to_label
```

```python
scores_baseline = pd.read_csv('scores_baseline.csv', index_col=0)
scores_resnet = pd.read_csv('scores_resnet.csv', index_col=0)
scores_mobilenet = pd.read_csv('scores_mobilenet.csv', index_col=0)

groups = scores_mobilenet.index.to_list()

df = pd.DataFrame({'custom': scores_baseline['disparate_impact'], 'resnet50': scores_resnet['disparate_impact'], 'mobilenet': scores_mobilenet['disparate_impact']},index=groups)

df = df.rename(index=dict(zip(groups, group_to_label(groups))))
#df = df.reindex(index=df.index[::-1])
```

```python
plt.figure()
ax = df.plot.bar(figsize=(10,6), title='Disparate impact comparison', xlim=(0.6, 1.2))
plt.savefig('modelcomp.png')
```

```python
scores_baseline = pd.read_csv('scores_baseline.csv', index_col=0)
scores_rw25 = pd.read_csv('scores_reweigh25.csv', index_col=0)
scores_rw50 = pd.read_csv('scores_reweigh50.csv', index_col=0)

groups = scores_baseline.index.to_list()

df = pd.DataFrame({'baseline': scores_baseline['disparate_impact'], '1.25 / 0.75': scores_rw25['disparate_impact'], '1.5 & 0.5': scores_rw50['disparate_impact']},index=groups)

df = df.rename(index=dict(zip(groups, group_to_label(groups))))
#df = df.reindex(index=df.index[::-1])
```

```python
plt.figure()
ax = df.plot.bar(figsize=(10,6), title='Disparate impact comparison', xlim=(0.6, 1.2))
plt.savefig('modelcomp.png')
```

```python

scores_resnet = pd.read_csv('scores_resnet.csv', index_col=0)
scores_resnet28 = pd.read_csv('scores_resnet28.csv', index_col=0)

groups = scores_resnet.index.to_list()

df = pd.DataFrame({'resnet 224x224': scores_resnet['disparate_impact'], 'resnet 28x28': scores_resnet28['disparate_impact']},index=groups)

df = df.rename(index=dict(zip(groups, group_to_label(groups))))
df = df.reindex(index=df.index[::-1])
```
